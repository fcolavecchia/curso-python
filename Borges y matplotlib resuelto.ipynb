```python
f = open("El Sur", "r")
texto = f.read()
print(texto)
```

```python
def borra_puntuacion(texto,puntuaciones):
    nuevo_texto = texto
    for signo in puntuaciones:
        nuevo_texto = nuevo_texto.replace(signo,' ')
            
    return nuevo_texto        
```

```python
puntuaciones = [',',';','.','\n','-','¿','?',':','(',')']
```

```python
texto_limpio = borra_puntuacion(texto,puntuaciones).lower()
print(texto_limpio)
```

```python
f = open('stopwords','r') 
stopwords = f.read().replace(' ','').split(',')
print(stopwords)
print(type(stopwords))
```

```python
def borra_stopwords(texto,stopwords):
    nuevo_texto = texto
    for stopword in stopwords:
        nuevo_texto = nuevo_texto.replace(' '+ stopword +' ',' ')
            
    return nuevo_texto        
```

```python
nuevo_texto = borra_stopwords(texto_limpio,stopwords)
print(nuevo_texto)
```

```python
lista_unica = list(set(nuevo_texto.split(' ')))
print(len(lista_unica))
print(lista_unica)
```

```python
def histograma_palabras(lista_unica,texto):
    resultado = {}
    """
    Dada una lista única de palabras y un texto,
    se genera un diccionario conteniendo cada palabra, y la frecuencia
    con la que aparece en el texto
    """
    todas = texto.split(' ')
    for palabra in lista_unica:
        resultado[palabra] = todas.count(palabra)
        
    return resultado    
        
    
```

```python
frec = histograma_palabras(lista_unica,nuevo_texto)
print(frec)
```

```python
wc = WordCloud(background_color="white", max_words=100)
# generate word cloud
wc.generate_from_frequencies(frec)

# show
plt.imshow(wc, interpolation="bilinear")
plt.axis("off")
plt.show()
```

```python
# Generate a word cloud image
from wordcloud import WordCloud
wordcloud = WordCloud(background_color="white", max_words=100).generate(nuevo_texto)

# Display the generated image:
# the matplotlib way:
import matplotlib.pyplot as plt
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
```
